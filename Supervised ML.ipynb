Importing Libaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression as SkLinearRegression
```

Loading Data

```python
data = pd.read_csv("Student - Score.csv")
print("size:",data.size,"; shape",data.shape)
data.head()
```

Plot Function

```python
def regression_plot(X,y,model,title=""):
  plt.figure(figsize=(14,7))
  plt.title(title)
  plt.xlabel("Hours")
  plt.ylabel("Scores")
  x_line = np.array([np.min(X) ,np.max(X) ]).reshape(-1,1)
  y_line = model.predict(x_line)
  plt.scatter(X, y,c='orange', label='Original Data Points')
  plt.plot(x_line, y_line,linewidth=4, label='Regression Line')
  plt.legend()
```

Processing The Data

```python
x = data['Hours'].values
y = data['Scores'].values
```

```python
X1 = x.reshape(-1,1)
```

Fitting and Plotting

```python
sk_lin_reg_model = SkLinearRegression().fit(X1, y)
regression_plot(X1,y,sk_lin_reg_model,title="Linear Regression Using Scikit Learn")
```

Problem : what will be predicted score if a student studies for 9.25 hrs/day?

```python
new_value = np.array([9.25])
new_value = new_value.reshape(-1,1)
prediction = sk_lin_reg_model.predict(new_value)
print(f"Prediction: The student get: {prediction[0]:.2f} marks if he studies 9.25 hrs")
```
